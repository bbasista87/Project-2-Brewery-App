```python
import requests
import json
import pandas as pd
import csv
import os
```

```python
beers_data= "beers-cleaned.csv"

beers_df = pd.read_csv(beers_data, encoding = "ISO-8859-1", engine='python')
beers_df.head()
```

```python
beer_df = beers_df.loc[:, ["brewery_id", "name", "style_id", "abv", "ibu"]]
beer_df.head()
```

```python
breweries_data= "breweries.csv"

breweries_df = pd.read_csv(breweries_data, encoding = "ISO-8859-1", engine='python')
breweries_df = breweries_df.rename(columns={'id': "brewery_id", "name": "brewery_name"})
breweries_df.head()
```

```python
breweries_df.drop(['address2', 'code', 'filepath', 'last_mod'], axis=1, inplace=True)                          
breweries_df.head()
```

```python
beer_df = beer_df.rename(columns={'name': "beer_name"})
beer_df.head()
```

```python
styles_data= "styles.csv"

styles_df = pd.read_csv(styles_data, encoding = "ISO-8859-1", engine='python')
styles_df.head(2)
```

```python
styles_df = styles_df.loc[:, ["id","style_name", "Description"]]
styles_df.head(2)
```

```python
beer_styles = pd.merge(beer_df, styles_df, left_on="style_id", right_on="id")
beer_styles.head()
```

```python
beerdf = pd.merge(beer_styles, breweries_df, on="brewery_id")
beerdf.head()
```

```python
beerdf = beerdf.loc[:, ["beer_name", "abv", "ibu","style_name", "brewery_name"]]
beerdf
```

```python
beerdf.dtypes
```

```python
beerdf = beerdf.rename(columns={'brewery_name': 'brewery','beer_name':'beer','style_name':'style'})
beerdf.columns.values
```

```python
beerdf.to_csv('beers_final.csv', index=False)
```

```python
geo_data= "breweries_geocode.csv"

geo_df = pd.read_csv(geo_data, encoding = "ISO-8859-1", engine='python')
geo_df.head(2)
```

```python
brewerydf = pd.merge(breweries_df, geo_df, on="brewery_id")
brewerydf.head(2)
```

```python
brewerydf = brewerydf.loc[:, ["brewery_name", "address1", "city", "state","latitude","longitude","website"]]
brewerydf
```

```python
brewerydf.dtypes
```

```python
brewerydf = brewerydf.rename(columns={'brewery_name': 'brewery','address1':'address'})
brewerydf.columns.values
```

```python
brewerydf.to_csv('breweries_final.csv', index=False)
```
