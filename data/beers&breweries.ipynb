```python
import requests
import json
import pandas as pd
import csv
import os
```

```python
beers_data= "beers-cleaned.csv"

beers_df = pd.read_csv(beers_data, encoding = "ISO-8859-1", engine='python')
beers_df.head()
```

```python
beer_df = beers_df.loc[:, ["brewery_id", "name", "style_id", "abv", "ibu"]]
beer_df.head()
```

```python
breweries_data= "breweries.csv"

breweries_df = pd.read_csv(breweries_data, encoding = "ISO-8859-1", engine='python')
breweries_df = breweries_df.rename(columns={'id': "brewery_id", "name": "brewery_name"})
breweries_df.head()
```

```python
breweries_df.drop(['address2', 'code', 'filepath', 'last_mod'], axis=1, inplace=True)                          
breweries_df.head()
```

```python
beer_df = beer_df.rename(columns={'name': "beer_name"})
beer_df.head()
```

```python
styles_data= "styles.csv"

styles_df = pd.read_csv(styles_data, encoding = "ISO-8859-1", engine='python')
styles_df.head(2)
```

```python
styles_df = styles_df.loc[:, ["id","style_name", "Description"]]
styles_df.head(2)
```

```python
beer_styles = pd.merge(beer_df, styles_df, left_on="style_id", right_on="id")
beer_styles.head()
```

```python
brewery_beer = pd.merge(breweries_df, beer_styles, on="brewery_id")
brewery_beer.head(2)                    
```

```python
geo_data= "breweries_geocode.csv"

geo_df = pd.read_csv(geo_data, encoding = "ISO-8859-1", engine='python')
geo_df.head(2)
```

```python
beerdf = pd.merge(brewery_beer, geo_df, on="brewery_id")
beerdf.head(2)
```

```python
beerdf = beerdf.loc[:, ["brewery_name","beer_name", "abv", "ibu","style_name", "address1", "city", "state","latitude","longitude","website"]]
beerdf
```

```python
beerdf.to_csv('breweries.csv', index=False)
```
